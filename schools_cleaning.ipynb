```python
# Dependencies
import pandas as pd

# Read data into Pandas DataFrame
file = "Public_Schools_2020-2021_Cleaned.csv"
df = pd.read_csv(file)
```

```python
# Reduce to desired columns
red_df = df[["SCH_NAME", "LSTREET1", "LCITY", "LSTATE", "LZIP", "VIRTUAL","TOTFRL","TOTAL", "WH", "FTE", "LATCOD", "LONCOD", "ULOCALE", "STUTERATIO","SCHOOL_TYPE_TEXT", "SY_STATUS_TEXT", "SCHOOL_LEVEL"]]

# Ensure the following columns are numeric: LZIP, TOTFRL, TOTAL, WH, FTE, LATCOD, LONCOD, STUTERATIO
red_df[["LZIP", "TOTFRL", "TOTAL", "WH", "FTE", "LATCOD", "LONCOD", "STUTERATIO"]] = red_df[["LZIP", "TOTFRL", "TOTAL", "WH", "FTE", "LATCOD", "LONCOD", "STUTERATIO"]].apply(pd.to_numeric) #errors = 'ignore'

# Remove rows where there is missing data
new_df = red_df.dropna()

# View some of new_df
new_df.head()
```

```python
# Look at how many rows we have lost with the removal of empty rows
# Final result
new_df.describe()
```

```python
# Dataset before removal of empty data
red_df.describe()

```

```python
# Filter by state of Maryland
new_df = new_df.loc[new_df["LSTATE"] == "MD", :]
new_df.head()
```

```python
# Filter by "Regular school"
new_df = new_df.loc[new_df["SCHOOL_TYPE_TEXT"] == "Regular school", :]
new_df
```

100,722 - 73,577 = 27,145 rows/schools removed.

```python
# Save new df
new_df.to_csv("Maryland_Public_Schools_2020-2021_Cleaned.csv")
```
